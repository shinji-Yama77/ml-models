```python
original_img = plt.imread('bird_small.png')
```

```python
# For visualizing the colors

# this gives the RGB which have three values (R, G, B) each ranging from 0-255
# intensity. Here we plot the colors to visualize the shape

rgb_values = original_img[50, 33, :] 
                                     

# Create a 1x1 grid plot
fig, ax = plt.subplots(1, 1, figsize=(3, 3))

# Display the color using a patch
ax.add_patch(plt.Rectangle((0, 0), 1, 1, color=rgb_values))
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```

```python
X_img = np.reshape(original_img, (original_img.shape[0] * original_img.shape[1], 3))
```

```python
X_img[0]
```

```python
K=10
K1 = 30
km1 = KMeans(10, clusters=K)
km2 = KMeans(10, clusters=K1)
        
```

```python
centroids, idx = km1.fit(X_img)
```

```python
# For visualizing the colors

# this gives the RGB which have three values (R, G, B) each ranging from 0-255
# intensity. Here we plot the colors to visualize the shape
rgb_values = centroids[:, :]

# Determine the number of centroids and colors
num_colors = rgb_values.shape[0]

# Create a 1xn grid plot, where n is the number of colors
fig, ax = plt.subplots(1, num_colors, figsize=(num_colors, 1))

# Display each color using a patch
for i in range(num_colors):
    color = rgb_values[i]
    ax[i].add_patch(plt.Rectangle((0, 0), 1, 1, color=color))
    ax[i].set_xticks([])
    ax[i].set_yticks([])

plt.show()
```

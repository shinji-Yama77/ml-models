```python
import numpy as np
import matplotlib.pyplot as plt
from kmeans import KMeans
```

```python
original_img = plt.imread('neko-100.png')
```

```python
# visualizing original image


plt.imshow(original_img) # original image
```

```python
# What are it's dimensions

original_img.shape
```

```python
# For visualizing the colors

# this gives the RGB which have three values (R, G, B) each ranging from 0-255
# intensity. Here we plot the colors to visualize the shape

rgb_values = original_img[50, 33, :] 
                                     

# Create a 1x1 grid plot
fig, ax = plt.subplots(1, 1, figsize=(3, 3))

# Display the color using a patch
ax.add_patch(plt.Rectangle((0, 0), 1, 1, color=rgb_values))
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```

```python
X_img = np.reshape(original_img, (original_img.shape[0] * original_img.shape[1], 4))
```

```python
X_img[0]
```

```python
K = 16
K1 = 32
km1 = KMeans(10, clusters=K)
km2 = KMeans(10, clusters=K1)
```

```python
centroids, idx = km1.fit(X_img)
```

```python
centroids1, idx1 = km2.fit(X_img)
```

```python
X_recovered = centroids[idx, :] 


# Reshape image into proper dimensions
X_recovered = np.reshape(X_recovered, original_img.shape) 
```

```python
X_1 = centroids1[idx1, :]
X_1 = np.reshape(X_1, original_img.shape)
```

```python
fig, ax = plt.subplots(1,3, figsize=(16,16))
plt.axis('off')

ax[0].imshow(original_img)
ax[0].set_title('Original')
ax[0].set_axis_off()


# Display compressed image
ax[1].imshow(X_recovered)
ax[1].set_title('Compressed with %d colours'%K)
ax[1].set_axis_off()


ax[2].imshow(X_1)
ax[2].set_title('Compressed with %d colours'%K1)
ax[2].set_axis_off()
```
